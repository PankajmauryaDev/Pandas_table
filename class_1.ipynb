# series in Pandas

```python
import pandas as pd
```

```python
a = pd.Series([1,2,3,4,5,6,7])
print(a)
print(type(a))
print(a[2])
```

# properties of series

```python
import pandas as pd
```

```python
# Homogenous

s = pd.Series([1,2,3,4,5,"abj"])
print(s)
```

```python
# vectorized operation

s1 = pd.Series([1,2,3,5,6])
s2 = pd.Series([12,23,54,23])

print(s1+s2)
```

# DataFrame

```python
import pandas as pd
```

```python
# from distrinary
info = {
    "Name" : ["adam","bob","charli","alexa"],
    "age" : [40,23,12,23],
    "gpa" : [2.4,5.7,9.7,6.9]
}
# print(info["Name"])
# print(info)

df = pd.DataFrame(info)


print(df)
```

```python
print(df.index)
print(df.columns)
```

```python
# from list to list

lst = pd.DataFrame([[12 , "pankaj"],[32 , "maurya"], [45 , "rajan"]] , columns = ["roll","name"])
print(lst)
```

```python
# from numpy array list

import numpy as np

arr = np.array([[1,2,3] , [4,5,6]])
print(arr)

df = pd.DataFrame(arr , columns = ["a" , "b" , "c"])
print(df)
```
